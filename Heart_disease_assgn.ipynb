```python
from pandas.api.types import is_numeric_dtype
```

```python
for col in X.columns:
  if is_numeric_dtype(X[col]):
    print('%s:'%(col))
    print('\t Mean = %.2f'% X[col].mean())
    print('\t Standard Deviation=%.2f'%X[col].std())
    print('\t Minimum Value =%.2f'%X[col].min())
    print('\t Maximum Value =%.2f'%X[col].max())
```
